```python
def read(input1,ml1=None,output1=None):
    # read ml1 and assign index compatible to Sam's file
    cols=['Confidence level (based on deviation from the theorethical CE time)',
          'Probability to be present in urine_MP']
    dfml1=pd.read_excel(ml1)
    dfml1.index=dfml1['ID ML1'].str[1:].astype(int)
    dfml1=dfml1[cols]
    
    # create output xlsx for storing the modified sheets
    writer = pd.ExcelWriter(output1, engine='xlsxwriter')
    
    # read Sams' analyse per sheet
    xl=pd.ExcelFile(input1)
    #print (xl.sheet_names)
    for i in xl.sheet_names:
        print (i)
        df1=xl.parse(i,index_col=0,usecols=cols2)
        df1=df1.join(dfml1)
        print (df1.head())
        df1.to_excel(writer,sheet_name=i)
        #break
        
    # save the excel file to memory
    writer.save()
    
read(fsam+'merged_UPTODATE_v2_with_egfr20_ForHarald.xlsx',
    f1+'ML1 02122020 semi-final_v3_sequenced.xlsx',
    fsam+'merged_UPTODATE_v2_with_egfr20_ForHarald_confidence.xlsx')
    
```
