```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

import seaborn as sns; sns.set()
import scipy.stats as stats
import statsmodels.api as sm

from set_path import *
```

```python
class LR:
    def __init__(self, inputpep,inputpat,var,output1,outputresid):
        self.inputpep=inputpep
        self.inputpat=inputpat
        self.output1=output1
        self.outputresid=outputresid
        self.var=var
    def read(self):
        
        self.dfpat=pd.read_csv(self.inputpat,index_col=0)[self.var]
        
#         print (self.dfpat.columns.tolist())
#         print (self.dfpat.head())    
        self.dfpep=pd.read_csv(self.inputpep,index_col=0)
    
    def join(self):
        self.dfpat=self.dfpat.reindex(self.dfpep.index)
#         print (self.dfpep.head())
#         print (self.dfpat.head())  
        
#         print (self.df1.shape)
#         print (self.df1.head())



    def ols(self):
        npep=len(self.dfpep.columns)
        npat=len(self.dfpat)
        
        # fill na with 0
        self.dfpat.fillna(0, inplace=True)

        slp_ci=np.empty((npep,2))
        self.dfresid=pd.DataFrame(index=self.dfpat.index,
                           columns=self.dfpep.columns.tolist())
        self.df_out=pd.DataFrame(index=self.dfpep.columns.tolist(),
                           columns=['coef','std_err','pvalue'])
        
        # egfr, always the last column
        y=self.dfpat.iloc[:,-1]
        for i in range(npep):
#             print (self.dfpep.iloc[:,i][:,np.newaxis].shape)
#             print (self.dfpat.iloc[:,:-1].shape)
            x=np.concatenate((self.dfpep.iloc[:,i][:,np.newaxis],self.dfpat.iloc[:,:-1]),
                             axis=1)\
                .reshape(npat,-1)
            
#             print (x.shape,y.shape)
#             print (x)
#             print (y)
            
            x=sm.add_constant(x)
#             print (x.shape, y.shape)
            lr=sm.OLS(y,x)
            res=lr.fit()
        
            # res.resid is the residues y.fit- ytrue of each peptides
            # res.params[1] is the coefficient of the 1st parameter, i.e. peptide
            # res.bse[1] is the pvalue of the 1st parameter, i.e. peptide
            self.dfresid.iloc[:,i]=res.resid
            self.df_out.iloc[i]=[res.params[1],res.bse[1],res.pvalues[1]]
#             print (res.params)
#             print (res.pvalues)
#             break
        
        self.df_out=self.df_out.sort_values(['pvalue'], ascending = False)
    #Formula of Qvalue refer to the excel file in computerlab1
        Qvalue = [1]
        for row_num in range (len(self.df_out)):
    #important, q-value is the minimum FDR, so it always compares it with the preceding term, her the last term in the Qvalue list.
            q = min(len(self.df_out)*self.df_out.iloc[row_num]['pvalue']/(len(self.df_out)-row_num),Qvalue[-1])
            Qvalue.append(q)

#to remove the first term (1) from the list of q-value
        Qvalue.remove(1)
        self.df_out['pvalue_adj'] = Qvalue
#         print (self.df_out.head())
#         print (self.dfresid)
        
    
    def export_results(self):
        self.df_out.to_csv(self.output1)
        self.dfresid.to_csv(self.outputresid)
```

```python
x1=LR(inputpep=f2+'reduced_all_pivot.csv',
      inputpat=f2+'Tianlin_GFR_CKD_EPI_cleaned_20201005.csv',
      var=[
          #'SampleID', 
           'gender', 
           'Age', 
           'Diabetes/Obesity/metabolic syndrome',
#            'Chronic kidney disease', 
           'CVD/Hypertension',
           'GFR_CKD_EPI', ],
      output1=f2+'all_out.csv',
     outputresid=f2+'all_resid.csv')
x1.read()
x1.join()
x1.ols()
x1.export_results()
```
